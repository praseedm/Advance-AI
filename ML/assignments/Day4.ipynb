### 1. Prepare an ML model using KMeans algorithm to cluster some sample input
generated using make_blob function. Plot the clusters.

```python
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X,y_true = make_blobs(n_samples = 300, centers=4,cluster_std = 0.6)

#plt.scatter(X[:,0],X[:,1],s=50)
#plt.show()

model = KMeans(n_clusters = 4)
model.fit(X)

p = model.predict(X)

plt.scatter(X[:,0],X[:,1],s=50,c=p,cmap = 'viridis')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

print "Confusion Matrix\n",confusion_matrix(y_true,p)
print "Accuracy Score",accuracy_score(y_true,p) # These evaluation metrics not applicable for clustering
```

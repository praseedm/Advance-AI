## 1. Develop an ML model to predict the house price for the boston
housing dataset included in sklearn. Find out the RMSE values for
models developed using different algorithms and select the best one.
Also plot the scatter diagrams showing the actual and predicted values

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
```

```python
house_data = load_boston()
X = house_data.data
y = house_data.target
```

```python
print X.shape
print y.shape
```

#### Using Regression models
 * Linear regression
 * SVR
 * Random forest
 * Kneighbors Regressor
 * Ridge
 

```python
from sklearn.linear_model import Lasso
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge
 
```

```python
rmses = []
def test_model(m_name,model):
    global X,y,rmses
    model.fit(X,y)
    p = model.predict(X)
    rmses.append((m_name,(mean_squared_error(y,p) ** 0.5)))
    print "RMSE for ",m_name," : ",(mean_squared_error(y,p) ** 0.5)
    plt.scatter(y,p)
    plt.title(m_name)
```

```python
#Linear


test_model("Linear _ LASSO",Lasso())
```

```python
#SVR


test_model("SVR",SVR())
```

```python
#RandomForestRegressor
test_model("RandomForestRegressor",RandomForestRegressor())
```

```python
#KNeighborsRegressor
test_model("KNeighborsRegressor",KNeighborsRegressor())
```

```python
#Linear Ridge
test_model("Ridge",Ridge())
```

#### RMSE summary for all models

```python
rmses
```

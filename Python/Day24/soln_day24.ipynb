### 1.create a dataframe object from a (10,4)2D array  consisting of random integers between 10 and 20 by making c1,c2,c3,c4 as column indices.
### 2.Sort the above object based on the key 'c1' in ascending and then by 'c3' in descending order.

```python
import numpy as np
from pandas import DataFrame as df

s = np.arange(10,50)
np.random.shuffle(s)

df1 = df(s.reshape((10,4)),columns = 'c1 c2 c3 c4'.split())
```

```python
df1
```

```python
df1.sort_values('c3')
```

```python
df1.sort_values(['c1','c3'],ascending=[True,False])
```

### 	3.write script to store item,place and total sale in a dataframe object.There should be 3 or more places and 4 or more items in the set.
	     Based on the choice entered by user, 	
           * show placewise rank for a particular item (for a given item name).
	       * Show itemwise rank for a particular place (for a given place name).

```python
import random

item  = ['banana','orange','mango','grape']*5
place = ['nielit']*4+['kattangal']*4+['KA']*4+['TN']*4+['UP']*4
#random.shuffle(item)
#random.shuffle(place)
total = random.sample(range(20, 100), 20)

data = { 'item' : item, 'place' : place ,'tot_s' : total}
```

```python
ddf = df(data)
ddf
```

```python
# shuffled dataframe 
index = [i for i in range(ddf.shape[0])]
random.shuffle(index)
s_ddf =  ddf.set_index([index]).sort_index()
s_ddf
```

```python
pa = s_ddf[s_ddf['item'] == 'banana']
pa
```

```python
def item_summary(item):
    part = s_ddf[s_ddf['item'] == item]
    t = part['tot_s'] 
    rank = t.rank(ascending=False)
    #print rank
    part['rank'] = rank
    return part
```

```python
i =1
while(i != 0):
    i = input("------MENU------\n\t 1-Banana\n\t 2-Mango\n\t 3-Orange\n\t 4-Grape\n\t 0-EXIT")
    if(i == 1):
        print item_summary('banana'),'\n'
    elif(i == 2):
        print item_summary('mango') ,'\n' 
    elif(i == 3):
        print item_summary('orange'),'\n'
    elif(i == 4):
        print item_summary('grape') ,'\n'   
    elif(i == 0):
        print 'EXITING....'    
```

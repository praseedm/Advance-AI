### 1. Write a program to create and show different categories of employees having the following attributes.
     
        3.1 Employees are of two main categories Engineers and Officers.
        3.2 Each employee has a name,no,basic pay,da(30% of basic),hra(10% of basic) and sp.allowance.
	3.3 sp.allowance for engineers is 20% and 10% for Officers.
	3.4 Engineers are grouped into Junior Engrs and Senior Engrs.
        3.5 Junior Engrs will get an additional 500 Rupees with their hra.
        3.6 Senior Engrs will get an additional 1000 Rupees with their hra.
   	
	(hint:Engineer and Officer should inherit from Employee,JuniorEngr and SeniorEngr should inherit from Engineer,
          Instantiation of Employee and Engineer should be denied)

```python
import abc

class Employee:
    __metaclass__ = abc.ABCMeta
    no = 200
    def __init__(self):
        Employee.no += 1
        self.no = Employee.no
        self.name = raw_input("Name?")
        self.bpay = 0
        self.da = 0
        self.hra = 0
        self.spa = 0
        self.addi = 0
        self.salary = 0
        self.cat=""
    
    @abc.abstractmethod
    def get_bpay(self,bp):
        pass
    
    @abc.abstractmethod
    def get_spa(self,spa):
        pass 
    
    @abc.abstractmethod
    def get_addi(self,additional):
        pass   
        
    
    def cal_sal(self):
        self.da = 0.3* self.bpay
        self.hra = 0.1* self.bpay
        self.salary = self.bpay+self.da+self.hra+self.spa+self.addi
    
    def display(self):
        print "--info--\n Name :%s\n Category :%s\n No : %d\n Bpay : %d\n DA : %d\n HRA : %d\n SPA : %d\n Additional.A : %d\n T.Salary : %d"%(self.name,self.cat,self.no,self.bpay,self.da,self.hra,self.spa,self.addi,self.salary)
     

class Eng(Employee):
    def __init__(self):
        Employee.__init__(self)
        self.get_bpay(20000)
        self.get_spa(0.2*self.bpay)
     
    def get_bpay(self,bp):
        self.bpay = bp
    
    def get_spa(self,spa):
        self.spa = spa 
        
class Off(Employee):
    def __init__(self):
        Employee.__init__(self)
        self.cat = "Officer"
        self.get_bpay(10000)
        self.get_spa(0.1*self.bpay)
    def get_bpay(self,bp):
        self.bpay = bp
    
    def get_spa(self,spa):
        self.spa = spa     
        
class Jeng(Eng):
    def __init__(self):
        Eng.__init__(self)
        self.cat = "Jnr.Engineer"
        self.get_addi(500)
    def get_addi(self,additional):
        self.addi = additional     

class Seng(Eng):
    def __init__(self):
        Eng.__init__(self)
        self.cat = "Snr.Engineer"
        self.get_addi(1000)
    def get_addi(self,additional):
        self.addi = additional     

se = Jeng()
se.cal_sal()
se.display()
        
        
```

### 2. Modify the vector program(PythonLab_Day09_07Nov-QNo.01) to add equality checking (eg. v1==v2 should return True or False)
        (v1 and v2 are vector class instances)

```python
class vector:
    def __init__(self,px,py):
        self.x = px
        self.y = py
        
    def display(self):
        print "vector : (%.3f,%.3f)"%(self.x,self.y)
        
    def __add__(self,v2):
        return vector(self.x+v2.x, self.y+v2.y )
    
    def __eq__(self,v2):
        return (self.x == v2.x and self.y == v2.y)
 
v1 = vector(2,1)
v2 = vector(2,2)
print v1 == v2
```

### 3.Write a program(of your own choice) to show multiple inheritance. 

```python
class exam:
    def __init__(self):
        self.ename = raw_input("Subject?")
        self.emark = input("Mark?")

class lab:
    def __init__(self):
        self.lname = raw_input("Lab name?")
        self.lmark = input("Mark?")

class student(exam,lab):
    def __init__(self):
        self.sname = raw_input("Name?")
        exam.__init__(self)
        lab.__init__(self)
        self.tot = self.emark+self.lmark
    
    def display(self):
        print "--info--\n Name : %s\n %d in %s Lab\n %d in %s Exam\n Total : %d"%(self.sname,self.lmark,self.lname,self.emark,self.ename,self.tot)
        
s1 = student()
s1.display()
```

### 4.Write a program(of your own choice) to show multilevel inheritance.


```python
class vehicle:
    no=1001
    def __init__(self):
        vehicle.no += 2
        self.owner = raw_input("Owner?")
        self.wheels = 0
        self.vno = vehicle.no
        
class car(vehicle):
    def __init__(self):
        vehicle.__init__(self)
        self.cat = "Car"
        self.wheels = 4

class sp_car(car):
    def __init__(self):
        car.__init__(self)
        self.subc = "sports car"
    
    def display(self):
        print "--info--\n Owner:%s\n Vehicle no:%d\n No wheels:%d\n Type:%s\n Category:%s"%(self.owner,self.vno,self.wheels,self.cat,self.subc)
        
lamb = sp_car() 
lamb.display()
```

### 6. Write a program to print all the exception classes.
	the output should be as follows,
	BaseClass
	-subclass1
	--Subclass11
	--SubClass12
	---SubClass121
	----SubClass1211
	---SubClass122
	-Subclass2
	--SubClass21	
	Here SubClass12 is the base class for SubClass121  

```python
def subclass_info(cname,count=0):
    print '-'*count,cname.__name__
    for subc in cname.__subclasses__():
        subclass_info(subc,count+1)
        
subclass_info(Exception)        
```

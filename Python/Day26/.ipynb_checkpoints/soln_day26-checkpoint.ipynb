### 6.From the given csv files(items.csv and sales.csv in common/Python_Exercises folder) create two dataframe objects- idf1 and sdf1

```python
idf1 = pd.read_csv('items.csv')
sdf1 = pd.read_csv('sales.csv')
print idf1.head()
print sdf1.head()
```

### 7.Fill stock of Mobile phone with average stock. 

```python
idf1 = idf1.fillna(idf1.mean())
```

### 8.Using idf1 and sdf1 print item name, region and sales quantity and store the same in a new dataframe

```python
dfmg = idf1.merge(sdf1,left_on='id',right_on='tid')
dfmg
```

### 9.Create two dataframes by applying pivot on the above dataframe with region/item name as index/column.
